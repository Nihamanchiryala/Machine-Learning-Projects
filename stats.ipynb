```python
from itertools import permutations
count=0
data=[ 'a' , 'b' , 'c' ]
perm=permutations(data,2)
for i in list(perm):
     print(i)
     count+=1
print("permutation count: ",count)
```

#Combinations

```python
from itertools import combinations
count=0
data=[ 'a' , 'b' , 'c' ]
comb=combinations(data,2)
for i in list(comb):
     print(i)
     count+=1
print("combination count: ",count)
```

```python
from itertools import combinations_with_replacement as cr
count=0
data=[1,2,3]
combre=cr(data,2)
for i in list(combre):
      print(i)
      count+=1
print("combination with replacement: ",count)
```

```python
dataset = pd.read_csv("Dataset.csv")
print(dataset.shape)
print(dataset.describe())
print(dataset.head(5))
```

```python
simpleRandomSample = dataset.sample(n=10).sort_values(by="Id")
mean_simpleRandomSample=round(simpleRandomSample['SalePrice'].mean(),3)
print("mean_simpleRandomSample:",mean_simpleRandomSample)
print(simpleRandomSample)
```

```python
index = np.arange(0,len(dataset),step=25)
print(index)
systematicSample = dataset.iloc[index]
mean_systematicSample = round(systematicSample['SalePrice'].mean(),3)
print("mean_systematicSample :",mean_systematicSample)
```

```python
n=5
dataset['cluster_id']=np.repeat([range(1,n+1)],len(dataset)/n)
index=[]
for i in range(0,len(dataset)):
  if dataset['cluster_id'].iloc[i]%2==0:
    index.append(i)
clusterSample=dataset.iloc[index]
mean_clusterSample=round(clusterSample['SalePrice'].mean(),3)
print(mean_clusterSample)
```
