# Import Libraries

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("dataset.csv")
```

# Summarize Dataset

```python
print(dataset.shape)
print(dataset.head(5))
```

# Visualize Data

```python
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(dataset.area,dataset.price,color='black',marker="*")
```

# Segregating Data into X and Y

```python
X=dataset.drop("price",axis="columns")
X
```

```python
Y=dataset.price
Y
```

# Training

```python
model = LinearRegression()
model.fit(X,Y)
```

# Prediction

```python
x = 6660
LandAreainSqFt =[[x]]
PredictedModelResult =model.predict(LandAreainSqFt)
print(PredictedModelResult)
```

# Checking Our Model

## **Y=Mx+b**

```python
#calculating M
m=model.coef_
print(m)
```

```python
#calculate intercept
b=model.intercept_
print(b)
```

```python
# Y=mx+b
y = m*x + b
print("The price of {0} Square feet Land is {1}".format(x,y[0]))
```
