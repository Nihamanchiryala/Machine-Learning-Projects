# **Clustering using Income spent**

# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Loading Dataset

```python
from google.colab import files
uploaded=files.upload()
```

```python
dataset = pd.read_csv("dataset.csv")
```

# Summarize Data

```python
print(dataset.shape)
print(dataset.describe())
print(dataset.head(5))
```

# Segregate and Zipping Data

```python
Income = dataset['INCOME'].values
Spend = dataset['SPEND'].values
X=np.array(list(zip(Income,Spend)))
print(X)
```

# Find the Optimized K value

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
  km=KMeans(n_clusters = i,random_state = 0)
  km.fit(X)
  wcss.append(km.inertia_)
plt.plot(range(1,11),wcss,color='red',marker="+")
plt.title("Optimal K Value")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

# Finding the KMeans To Dataset

```python
model = KMeans(n_clusters=5,random_state=0)
y_means = model.fit_predict(X)
```

# **Visualizing the clusters**
# Cluster - 1 : Customers with medium income and Low spend
# Cluster - 2 : Customers with high income and medium to high spend
# Cluster - 3 : Customers with Low income 
# Cluster - 4 : Customers with medium income but High spend

```python
plt.scatter(X[y_means==0,0],X[y_means==0,1],s=50,marker="+",c="green",label="1")
plt.scatter(X[y_means==1,0],X[y_means==1,1],s=50,marker="+",c="blue",label="2")
plt.scatter(X[y_means==2,0],X[y_means==2,1],s=50,marker="+",c="red",label="3")
plt.scatter(X[y_means==3,0],X[y_means==3,1],s=50,marker="+",c="teal",label="4")
plt.scatter(X[y_means==4,0],X[y_means==4,1],s=50,marker="+",c="orange",label="5")
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],s=100,marker='s',c="black",label="Centroids")
plt.title("Income Spent Analysis")
plt.xlabel("Income")
plt.ylabel("Spent")
plt.legend()
plt.show()
```
