# Import Libraries

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

# Load Dataset from Directory

```python
from google.colab import files
uploaded=files.upload()
```

```python
dataset=pd.read_csv("dataset1.csv")
```

# Summarize Data

```python
print(dataset.shape)
print(dataset.head(5))
```

# Visualize Data

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(dataset.area,dataset.price)
```

# Segregate Data into X and Y

```python
X=dataset.drop('price',axis='columns')
X
```

```python
Y=dataset.price
Y
```

# Splitting Data into Train and Test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.20,random_state=0)
```

# Training Dataset using Linear Regression

```python
model=LinearRegression()
model.fit(X_train,y_train)
```

# Visualize Linear Regression results

```python
plt.scatter(X,Y,color="red",marker="*")
plt.plot(X,model.predict(X))
plt.title("Linear Regression")
plt.xlabel("Area")
plt.ylabel("Price")
plt.show()
```

# R2Score

```python
r2score=model.score(X_test,y_test)
print(r2score)
```

# Adjusted R score of model

```python
n=len(dataset)
p=len(dataset.columns)-1
adjr=1-(1-r2score)*(n-1)/(n-p-1)
print(adjr)
```

# Prediction

```python
x=6500
landSquareFt=[[x]]
result=model.predict(landSquareFt)
print(result)
```
