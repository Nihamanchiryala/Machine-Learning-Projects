# Importing Dataset

```python
dataset = pd.read_csv("dataset.csv")
print(dataset.shape)
print(dataset.head(5))
```

# Segreagating Data into input and output

```python
features =dataset.iloc[:,10].values
labels=dataset.iloc[:,1].values
print(labels)
```

# Removing the Special charecters

```python
processesed_features =[]
for sentence in range(0,len(features)):
      processesed_feature = re.sub(r'\W',' ',str(features[sentence]))
      processesed_feature = re.sub(r'\s+[a-zA-Z]\s+',' ',processesed_feature)
      processesed_feature = re.sub(r'\^[a-zA-Z]\s+',' ',processesed_feature)
      processesed_feature = re.sub(r'\s+',' ',processesed_feature,flags=re.I)
      processesed_feature = re.sub(r'^b\s+','',processesed_feature)
      processesed_feature = processesed_feature.lower()
      processesed_features.append(processesed_feature)
```

# Feature Extraction from text

```python
nltk.download("stopwords")
vectorizer = TfidfVectorizer(max_features=2500,min_df=7,max_df=0.8,stop_words=stopwords.words('english'))
```

# Splitting dataset into test and train

```python
X_train,X_test,y_train,y_test=train_test_split(processesed_features,labels,test_size=0.20,random_state=0)
```

# Loading Random Forest Algorithm

```python
text_classifier = RandomForestClassifier(n_estimators=200,random_state=0)
text_classifier=text_classifier.fit(X_train,y_train)
```
